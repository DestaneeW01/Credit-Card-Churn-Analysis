##**Create a New Column called attrition_binary**

###This column will map Existing Customer and Attrited Customer to binary

```python
#Map strings to numbers for easier analysis later
df["attrition_binary"] = df["attrition_flag"].map({
    "Existing Customer": 0,
    "Attrited Customer": 1
})

#Verify mapping
df[["attrition_flag", "attrition_binary"]].head()
```

```python
#Calculate percentages
counts = df["attrition_flag"].value_counts()
percentages = counts / counts.sum() * 100

#Plot bars with labels
sns.barplot(x=counts.index, y=counts.values)
for i, v in enumerate(counts.values):
    plt.text(i, v + max(counts.values)*0.01, f"{percentages[i]:.1f}%", ha='center')
plt.title("Churn Distribution (%)")
plt.show()
```

#Bivariate Analysis

```python
# ===============================
# NUMERICAL FEATURES VS CHURN
# ===============================

num_bivar_cols = [
    "months_inactive_12_mon",
    "total_trans_amt",
    "total_trans_ct",
    "credit_limit",
    "avg_utilization_ratio"
]

for col in num_bivar_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x="attrition_flag", y=col, data=df)
    plt.title(f"{col} vs Churn")
    plt.show()

    # Summary stats
    print(df.groupby("attrition_flag")[col].describe())

    #Insights
    if col == "months_inactive_12_mon":
        print("Insight: Customers with higher inactivity (3-4 months) have significantly higher churn. Inactivity is a key driver.")
    elif col == "total_trans_amt":
        print("Insight: Customers with lower total transaction amounts tend to churn more, while high spenders are retained.")
    elif col == "total_trans_ct":
        print("Insight: Customers with fewer transactions per year have higher attrition, indicating engagement matters.")
    elif col == "credit_limit":
        print("Insight: Customers with lower credit limits show slightly higher churn, but the effect is smaller than activity-related variables.")
    elif col == "avg_utilization_ratio":
        print("Insight: High credit utilization correlates with churn — customers maxing out credit are more at risk.")
```

```python
# ===============================
# CATEGORICAL FEATURES VS CHURN
# ===============================
cat_bivar_cols = [
    "education_level",
    "marital_status",
    "income_category",
    "card_category"
]

for col in cat_bivar_cols:
    churn_rate = df.groupby(col)["attrition_binary"].mean().sort_values(ascending=False)

    plt.figure(figsize=(6,4))
    churn_rate.plot(kind="bar")
    plt.ylabel("Churn Rate")
    plt.title(f"{col} vs Churn Rate")
    plt.show()

    # Insights
    if col == "education_level":
      print("Insight: Customers with 'Unknown' or 'Uneducated' education levels show slightly higher churn, while higher education groups are too small for strong conclusions.")
    elif col == "marital_status":
        print("Insight: Single and Unknown marital status customers churn more than Married customers.")
    elif col == "income_category":
        print("Insight: Lower income and Unknown income customers tend to have higher churn rates.")
    elif col == "card_category":
        print("Insight: Most churn comes from Blue cardholders due to their large population; higher-tier cards have too few customers for reliable conclusions.")
```

#Correlation with Churn Analysis

```python
# ==================================
# STEP 8: CORRELATION ANALYSIS
# ==================================

# Verify binary column exists
df[["attrition_flag", "attrition_binary"]].head()

# Create correlation matrix (numeric only)
corr = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(12,8))
sns.heatmap(
    corr,
    annot=False,
    cmap="coolwarm",
    center=0
)
plt.title("Correlation Matrix")
plt.show()


# ==================================
# FOCUS ON CHURN CORRELATIONS ONLY
# ==================================

# Extract correlations with churn
churn_corr = corr["attrition_binary"]

# Sort values (ascending)
churn_corr_sorted = churn_corr.sort_values()

# Print sorted correlations
print(churn_corr_sorted)


# ==================================
# IDENTIFY STRONGEST VARIABLES
# ==================================

# Remove self-correlation (attrition_binary = 1.0)
churn_corr_no_self = churn_corr_sorted.drop("attrition_binary")

# Top 3 positive correlations (highest values)
top_positive = churn_corr_no_self.tail(3)

# Top 3 negative correlations (lowest values)
top_negative = churn_corr_no_self.head(3)

print("Top Positive Correlations with Churn:")
print(top_positive)

print("\nTop Negative Correlations with Churn:")
print(top_negative)
```
